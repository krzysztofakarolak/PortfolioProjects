```python
# First let's import the packages we will use in this project
import pandas as pd
import numpy as np
import seaborn as sns

import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
import matplotlib
plt.style.use('ggplot')
from matplotlib.pyplot import figure

%matplotlib inline
matplotlib.rcParams['figure.figsize'] = (12,8)

pd.options.mode.chained_assignment = None

#pd.set_option('display.max.rows',8000)   <-- max.rows option in pandas
#pd.set_option('display.max.columns',20)
# Now we need to read in the data
df = pd.read_csv(r'C:\Users\Krzysia\Documents\PortfolioProjects\Python\movies.csv')
# and take a look at the data
df
```

```python
# We need to see if we have any missing data
# Let's loop through the data and see if there is anything missing

for col in df.columns:
    pct_missing = np.mean(df[col].isnull())
    print('{} - {}%'.format(col, round(pct_missing*100)))
```

```python
# Data Types for our columns
print(df.dtypes)
```

```python
#changing data types i columns into strings to get rid off NaNs
df['budget']=df['budget'].apply(lambda x: str(x))
df['gross']=df['gross'].apply(lambda x: str(x))
```

```python
#clean columns of NAN etc.
df['budget']=df['budget'].str.replace('nan','0.0')
df['gross']=df['gross'].str.replace('nan','0.0')
df
```

```python
df['budget']=df['budget'].str.replace(r'\.0','')
df['gross']=df['gross'].str.replace(r'\.0','')
df
```

```python
#change data types in columns
df['budget']=df['budget'].astype('int64')
df['gross']=df['gross'].astype('int64')
df
```

```python
#Correct year column
df['right_year']=df['released'].astype(str).str.extract(pat = '([0-9]{4})')
df
```

```python
# ordering data according to gross (descending)
df = df.sort_values(by=['gross'], inplace=False, ascending=False)
```

```python
#Dropping duplicates 
#without equal sign it will only show possible values in the column
#if you want to see all possibilities, you need to switch on max.row option in pandas
df['company'].drop_duplicates().sort_values(ascending=False)
```

```python
#checking/getting rid off duplicates
df.drop_duplicates()
df
```

```python
#budget high correlation   #company high correlation
# scatter plot with budget vs gross
plt.scatter(x=df['budget'],y=df['gross'])
plt.title('Budget vs Gross Earnings')
plt.xlabel('Budget for Film')
plt.ylabel('Gross Earnings')
plt.show()
```

```python
#Plot budget vs gross using seaborn
sns.regplot(x='budget',y='gross',data=df, scatter_kws={'color':'red'},line_kws={'color':'blue'})
```

```python
#correlation  
#methods: pearson (<-- by default), kendall, spearman, callable    #numeric_only=True <-- on False by default
df.corr(numeric_only=True)
```

```python
c_m=df.corr(numeric_only=True)
sns.heatmap(c_m,annot=True)
plt.title('Correlation Matrix for Numeric features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
df_n = df

for col_name in df_n.columns:
    if(df_n[col_name].dtype == 'object'):
        df_n[col_name]= df_n[col_name].astype('category')
        df_n[col_name] = df_n[col_name].cat.codes
        
df_n
```

```python
c_m2=df_n.corr(numeric_only=True)
sns.heatmap(c_m2,annot=True)
plt.title('Correlation Matrix All Features')
plt.xlabel('Movie Features')
plt.ylabel('Movie Features')
plt.show()
```

```python
corr_pairs = c_m2.unstack()         #to show all data turn on option max.rows in pandas at the beginning
corr_pairs
```

```python
sorted_pairs= corr_pairs.sort_values()     #to show all data turn on option max.rows in pandas at the beginning
sorted_pairs
```

```python
H_corr=sorted_pairs[(sorted_pairs)>0.5]
H_corr
```
