# Notatki 
Zadanie dotyczy wyceny opcji oraz zabezpieczenia portfela przy u≈ºyciu strategii replikujƒÖcej (hedgingu), czyli techniki polegajƒÖcej na budowaniu portfela z aktyw√≥w, kt√≥re majƒÖ na celu odzwierciedlenie zmian w warto≈õci opcji. W tym przypadku jest to europejska opcja kupna (call), a strategia zak≈Çada comiesiƒôcznƒÖ aktualizacjƒô.

**Tematyka**

Opcja kupna (call) daje w≈Ça≈õcicielowi prawo, ale nie obowiƒÖzek, do kupna okre≈õlonego aktywa po ustalonej cenie (cenie wykonania, $K$) w okre≈õlonym terminie.

Strategia replikujƒÖca ‚Äì Bank zabezpiecza portfel, kupujƒÖc aktywo bazowe (ùíú) i inwestujƒÖc w aktywo wolne od ryzyka (np. obligacje). Kluczowym elementem tej strategii jest delta hedging, kt√≥ry polega na zakupie lub sprzeda≈ºy odpowiedniej ilo≈õci aktywa bazowego w taki spos√≥b, aby zabezpieczyƒá siƒô przed zmianami cen opcji.

Model Blacka-Scholesa ‚Äì Wykorzystuje siƒô go do wyceny opcji, a do zabezpieczenia portfela istotnym parametrem jest delta opcji ($\Delta$), kt√≥ra m√≥wi, jak bardzo zmieni siƒô warto≈õƒá opcji w odpowiedzi na zmianƒô ceny aktywa bazowego.

Wz√≥r Blacka-Scholesa na cenƒô europejskiej opcji kupna to:
$$C(S_t,K,T-t)=S_tN(d_+)-Ke^{-r(T-t)}N(d_-), $$
gdzie:

$S_t$ - cena aktywa bazowego w momencie $t$

$K$ - cena wykonania opcji

$r$ - stopa wolna od ryzyka

$\sigma$ - zmienno≈õƒá cen aktywa

$T$ - termin wyga≈õniƒôcia

$t$ - moment inwestycji

$N(d_+)$, $N(d_-)$ - warto≈õci funkcji dystrybuanty standaryzowanego rozk≈ÇƒÖdu normalnego, gdzie
$$d_+=\frac{\ln{\frac{S_t}{K}}+(r+\frac{\sigma^2}{2})(T-t)}{\sigma\sqrt{T-t}},$$ 
oraz
$$d_-=d_+-\sigma \sqrt{T-t}.$$
‚ÄãMo≈ºliwe, ≈ºe bƒôdziemy musieli rozwa≈ºyƒá wska≈∫nik delta, czyli
$$\Delta_C=N(d_+).$$

# Python
Napiszmy funkcjƒô, kt√≥ra oblicza≈Çaby nam ceny europejskiej opcji kupna wed≈Çug wzoru Blacka-Scholesa.

Argumenty funkcji:

S: Cena aktualna aktywa

K: Cena wykonania opcji

T: Czas do wyga≈õniƒôcia opcji

r: Stopa wolna od ryzyka

sigma: Wsp√≥≈Çczynnik zmienno≈õci

```python
import numpy as np
from scipy.stats import norm
```

```python
def black_scholes(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N = norm.cdf
    return S*N(d1) - K*np.exp(-r*T)*N(d2)
```

Dane do obliczenia cen:

```python
S_0 = 115  # Cena poczƒÖtkowa aktywa
K = 120  # Cena wykonania opcji
T = 1  # Czas do wyga≈õniƒôcia opcji (w latach)
r = 0.02  # Stopa wolna od ryzyka
sigma = 0.15  # Wsp√≥≈Çczynnik zmienno≈õci
S_1 = 117  # Cena akcji po pierwszym miesiƒÖcu
```

Ceny opcji:

```python
C_0 = black_scholes(S_0, K, T , r, sigma)
C_0
```

```python
C_1 = black_scholes(S_1, K, T - 1/12, r, sigma)
C_1
```

Ich r√≥≈ºnica:

```python
C_1-C_0
```

Python oferuje nam bardzo du≈ºƒÖ dok≈Çadno≈õƒá, szczeg√≥lnie gdy chodzi o obliczenia logarytmiczne, wyk≈Çadnicze, czy warto≈õci dystrybuanty. Jak widaƒá z tego wyniku ju≈º nie wywnioskowaliby≈õmy odpowiedzi *D*.

A co gdyby≈õmy przeformu≈Çowali naszƒÖ funkcjƒô, tak by korzysta≈Ça z warto≈õci dystrybuanty do piƒôciu miejsc po przecinku, jak my korzystajƒÖc z tablic?

```python
def black_scholes_round(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d1=round(d1,2)
    d2 = d1 - sigma*np.sqrt(T)
    d2=round(d2,2)
    N_d1 = round(norm.cdf(d1),5)
    N_d2=round(norm.cdf(d2),5)
    return S*N_d1 - K*np.exp(-r*T)*N_d2
```

```python
C_0r = black_scholes_round(S_0, K, T , r, sigma)
C_0r
```

```python
C_1r = black_scholes_round(S_1, K, T - 1/12, r, sigma)
C_1r
```

```python
C_1r-C_0r
```

Tutaj wnioski by≈Çyby takie same.

# Projekt 1 (Symulacja cen opcji europejskich metodƒÖ Monte Carlo)
InspirujƒÖc siƒô zadaniem i jego danymi przeprowad≈∫my symulacjƒô cen aktywa.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Parametry opcji
S0 = 115  # cena poczƒÖtkowa aktywa
K = 120  # cena wykonania opcji
r = 0.02  # roczna stopa wolna od ryzyka
sigma = 0.15  # zmienno≈õƒá ceny aktywa
T = 1  # czas do wyga≈õniƒôcia opcji (1 rok)
n_simulations = 10000  # liczba symulacji
n_steps = 12  # liczba krok√≥w (miesiƒôczna aktualizacja)
```

```python
# Parametry symulacji
dt = T / n_steps  # krok czasowy

# Symulacja ≈õcie≈ºek cen aktywa
np.random.seed(42)  # dla powtarzalno≈õci wynik√≥w
S = np.zeros((n_simulations, n_steps + 1))
S[:, 0] = S0

for t in range(1, n_steps + 1):
    z = np.random.standard_normal(n_simulations)  # losowanie liczb z rozk≈Çadu normalnego
    S[:, t] = S[:, t-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)

# Zobacz kilka pierwszych symulowanych ≈õcie≈ºek
pd.DataFrame(S).head()
```

```python
# Warto≈õƒá opcji kupna w momencie wyga≈õniƒôcia
payoff = np.maximum(S[:, -1] - K, 0)

# Warto≈õƒá opcji teraz (dyskontowanie wyp≈Çat)
option_price = np.exp(-r * T) * np.mean(payoff)
print(f"Szacowana cena opcji: {option_price:.4f}")
```

```python
# Wykres przyk≈Çadowych ≈õcie≈ºek cen aktywa
plt.figure(figsize=(10, 6))
plt.plot(S.T[:, :10], lw=1.5)
plt.title("Symulowane ≈õcie≈ºki cen aktywa")
plt.xlabel("MiesiƒÖc")
plt.ylabel("Cena aktywa")
plt.grid(True)
plt.show()

# Histogram wyp≈Çat z opcji
plt.figure(figsize=(10, 6))
plt.hist(payoff, bins=50, color='blue', edgecolor='black')
plt.title("Rozk≈Çad wyp≈Çat z opcji")
plt.xlabel("Wyp≈Çata")
plt.ylabel("Czƒôsto≈õƒá")
plt.grid(True)
plt.show()
```
