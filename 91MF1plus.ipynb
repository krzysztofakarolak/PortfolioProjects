# Notatki 
Zadanie dotyczy wyceny opcji oraz zabezpieczenia portfela przy użyciu strategii replikującej (hedgingu), czyli techniki polegającej na budowaniu portfela z aktywów, które mają na celu odzwierciedlenie zmian w wartości opcji. W tym przypadku jest to europejska opcja kupna (call), a strategia zakłada comiesięczną aktualizację.

**Tematyka**

Opcja kupna (call) daje właścicielowi prawo, ale nie obowiązek, do kupna określonego aktywa po ustalonej cenie (cenie wykonania, $K$) w określonym terminie.

Strategia replikująca – Bank zabezpiecza portfel, kupując aktywo bazowe (𝒜) i inwestując w aktywo wolne od ryzyka (np. obligacje). Kluczowym elementem tej strategii jest delta hedging, który polega na zakupie lub sprzedaży odpowiedniej ilości aktywa bazowego w taki sposób, aby zabezpieczyć się przed zmianami cen opcji.

Model Blacka-Scholesa – Wykorzystuje się go do wyceny opcji, a do zabezpieczenia portfela istotnym parametrem jest delta opcji ($\Delta$), która mówi, jak bardzo zmieni się wartość opcji w odpowiedzi na zmianę ceny aktywa bazowego.

Wzór Blacka-Scholesa na cenę europejskiej opcji kupna to:
$$C(S_t,K,T-t)=S_tN(d_+)-Ke^{-r(T-t)}N(d_-), $$
gdzie:

$S_t$ - cena aktywa bazowego w momencie $t$

$K$ - cena wykonania opcji

$r$ - stopa wolna od ryzyka

$\sigma$ - zmienność cen aktywa

$T$ - termin wygaśnięcia

$t$ - moment inwestycji

$N(d_+)$, $N(d_-)$ - wartości funkcji dystrybuanty standaryzowanego rozkłądu normalnego, gdzie
$$d_+=\frac{\ln{\frac{S_t}{K}}+(r+\frac{\sigma^2}{2})(T-t)}{\sigma\sqrt{T-t}},$$ 
oraz
$$d_-=d_+-\sigma \sqrt{T-t}.$$
​Możliwe, że będziemy musieli rozważyć wskaźnik delta, czyli
$$\Delta_C=N(d_+).$$

# Python
Napiszmy funkcję, która obliczałaby nam ceny europejskiej opcji kupna według wzoru Blacka-Scholesa.

Argumenty funkcji:

S: Cena aktualna aktywa

K: Cena wykonania opcji

T: Czas do wygaśnięcia opcji

r: Stopa wolna od ryzyka

sigma: Współczynnik zmienności

```python
import numpy as np
from scipy.stats import norm
```

```python
def black_scholes(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d2 = d1 - sigma*np.sqrt(T)
    N = norm.cdf
    return S*N(d1) - K*np.exp(-r*T)*N(d2)
```

Dane do obliczenia cen:

```python
S_0 = 115  # Cena początkowa aktywa
K = 120  # Cena wykonania opcji
T = 1  # Czas do wygaśnięcia opcji (w latach)
r = 0.02  # Stopa wolna od ryzyka
sigma = 0.15  # Współczynnik zmienności
S_1 = 117  # Cena akcji po pierwszym miesiącu
```

Ceny opcji:

```python
C_0 = black_scholes(S_0, K, T , r, sigma)
C_0
```

```python
C_1 = black_scholes(S_1, K, T - 1/12, r, sigma)
C_1
```

Ich różnica:

```python
C_1-C_0
```

Python oferuje nam bardzo dużą dokładność, szczególnie gdy chodzi o obliczenia logarytmiczne, wykładnicze, czy wartości dystrybuanty. Jak widać z tego wyniku już nie wywnioskowalibyśmy odpowiedzi *D*.

A co gdybyśmy przeformułowali naszą funkcję, tak by korzystała z wartości dystrybuanty do pięciu miejsc po przecinku, jak my korzystając z tablic?

```python
def black_scholes_round(S, K, T, r, sigma):
    d1 = (np.log(S/K) + (r + 0.5*sigma**2)*T) / (sigma*np.sqrt(T))
    d1=round(d1,2)
    d2 = d1 - sigma*np.sqrt(T)
    d2=round(d2,2)
    N_d1 = round(norm.cdf(d1),5)
    N_d2=round(norm.cdf(d2),5)
    return S*N_d1 - K*np.exp(-r*T)*N_d2
```

```python
C_0r = black_scholes_round(S_0, K, T , r, sigma)
C_0r
```

```python
C_1r = black_scholes_round(S_1, K, T - 1/12, r, sigma)
C_1r
```

```python
C_1r-C_0r
```

Tutaj wnioski byłyby takie same.

# Projekt 1 (Symulacja cen opcji europejskich metodą Monte Carlo)
Inspirując się zadaniem i jego danymi przeprowadźmy symulację cen aktywa.

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Parametry opcji
S0 = 115  # cena początkowa aktywa
K = 120  # cena wykonania opcji
r = 0.02  # roczna stopa wolna od ryzyka
sigma = 0.15  # zmienność ceny aktywa
T = 1  # czas do wygaśnięcia opcji (1 rok)
n_simulations = 10000  # liczba symulacji
n_steps = 12  # liczba kroków (miesięczna aktualizacja)
```

```python
# Parametry symulacji
dt = T / n_steps  # krok czasowy

# Symulacja ścieżek cen aktywa
np.random.seed(42)  # dla powtarzalności wyników
S = np.zeros((n_simulations, n_steps + 1))
S[:, 0] = S0

for t in range(1, n_steps + 1):
    z = np.random.standard_normal(n_simulations)  # losowanie liczb z rozkładu normalnego
    S[:, t] = S[:, t-1] * np.exp((r - 0.5 * sigma**2) * dt + sigma * np.sqrt(dt) * z)

# Zobacz kilka pierwszych symulowanych ścieżek
pd.DataFrame(S).head()
```

```python
# Wartość opcji kupna w momencie wygaśnięcia
payoff = np.maximum(S[:, -1] - K, 0)

# Wartość opcji teraz (dyskontowanie wypłat)
option_price = np.exp(-r * T) * np.mean(payoff)
print(f"Szacowana cena opcji: {option_price:.4f}")
```

```python
# Wykres przykładowych ścieżek cen aktywa
plt.figure(figsize=(10, 6))
plt.plot(S.T[:, :10], lw=1.5)
plt.title("Symulowane ścieżki cen aktywa")
plt.xlabel("Miesiąc")
plt.ylabel("Cena aktywa")
plt.grid(True)
plt.show()

# Histogram wypłat z opcji
plt.figure(figsize=(10, 6))
plt.hist(payoff, bins=50, color='blue', edgecolor='black')
plt.title("Rozkład wypłat z opcji")
plt.xlabel("Wypłata")
plt.ylabel("Częstość")
plt.grid(True)
plt.show()
```
